```python
import networkx as nx
import matplotlib.pyplot as plt
from tqdm import trange
```

```python
N = 1000
k = 5
```

```python
ap = []

for L in trange(1, 11):
    ap_subtotal = 0.
    for _ in range(100):
        G = [nx.erdos_renyi_graph(N, k / (N-1)) for _ in range(L)]
        aps = set(nx.articulation_points(G[0]))
        for g in G[1:]:
            aps = aps.union(nx.articulation_points(g))
        ap_subtotal += len(aps) / (N*L)
    ap.append(ap_subtotal / 100)
```

```python
plt.plot(range(1, 11), ap, 'o')
plt.xlabel('$L$')
plt.ylabel('$n_{AP}$')
```
