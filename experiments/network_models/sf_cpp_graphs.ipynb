```python
import json
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('graph_3.json', mode='r') as f:
    G1 = json.load(f)

with open('graph_3.5.json', mode='r') as f:
    G2 = json.load(f)

with open('graph_2.5.json', mode='r') as f:
    G3 = json.load(f)
```

```python
len(G1), len(G2), len(G3)
```

```python
degrees1 = [len(neighb) for _, neighb in G1.items()]
degrees2 = [len(neighb) for _, neighb in G2.items()]
degrees3 = [len(neighb) for _, neighb in G3.items()]
mean_degree1 = np.mean(degrees1)
mean_degree2 = np.mean(degrees2)
mean_degree3 = np.mean(degrees3)
mean_degree1, mean_degree2, mean_degree3
```

```python
_, bins, _ = plt.hist(degrees1, bins=np.logspace(0, 3), density=True, log=True)
plt.plot(bins, 100 / bins**3)
plt.xscale('log')
plt.yscale('log')
```

```python
_, bins, _ = plt.hist(degrees2, bins=np.logspace(0, 3), density=True, log=True)
plt.plot(bins, 100 / bins**3.5)
plt.xscale('log')
plt.yscale('log')
```

```python
_, bins, _ = plt.hist(degrees3, bins=np.logspace(0, 4), density=True, log=True)
plt.plot(bins, 100 / bins**2.5)
plt.xscale('log')
plt.yscale('log')
```
