```python
import os
import sys

import matplotlib.pyplot as plt
import numpy as np

if os.path.abspath('..') not in sys.path:
    sys.path.append(os.path.abspath('..'))
from Main import plot_multiplex_with_ap_removal
```

```python
plt.rc('text', usetex=True)
plt.rc('font', size=22, **{'family': 'sans-serif', 'sans-serif': ['Helvetica']})
plt.rcParams['xtick.major.pad'] = 8
plt.rcParams['ytick.major.pad'] = 8
```

```python
N = 5000
T = .4
gamma = 3.
c = 8.
```

```python
g_vals = [.8, .9, 1.]
# kbar_vals = [5., 6., 7., 8., 9., 10.]
nu_vals = [0., 1.]
```

```python
fig = plt.figure(
    # figsize=(10, 10),
    # layout='constrained',
)
subfigs = fig.subfigures(nrows=len(g_vals), ncols=len(nu_vals),
                         wspace=0.,
                         squeeze=False)
# fig, axes = plt.subplots(
#     nrows=len(g_vals),
#     ncols=len(kbar_vals),
#     figsize=(12, 12),
#     subplot_kw={'projection': 'polar'},
# )

DISC_WIDTH = .45  # relative

for i, g in enumerate(g_vals):
    for j, nu in enumerate(nu_vals):

        print(f'\n{g = }, {nu = }\n')
        subfig = subfigs[i, j]
        # subfig.set_facecolor('0.75')

        # if i == len(g_vals) - 1:
            # ax = subfig.add_subplot(111, frameon=False)
            # ax.set_xlabel(fr'$c = {kbar}$')
            # ax.spines['top'].set_color('none')
            # ax.spines['bottom'].set_color('none')
            # ax.spines['left'].set_color('none')
            # ax.spines['right'].set_color('none')
            # ax.tick_params(labelcolor='w', top=False, bottom=False, left=False, right=False)
            # ax.set_xlabel(fr'$c = {kbar}$')

        ax1, ax2 = subfig.subplots(1, 2, subplot_kw={'projection': 'polar'})
        plot_multiplex_with_ap_removal(N, gamma, c, T, nu, g, ax1, ax2, seed=42)
        # subfig.subplots_adjust(left=0, right=1, bottom=0, top=1, wspace=0)

        ax1.set_xlabel('', labelpad=0)
        ax1.set_ylabel('', labelpad=0)
        ax2.set_xlabel('', labelpad=0)
        ax2.set_ylabel('', labelpad=0)

        ax1.set_position([.5 - DISC_WIDTH, 0, DISC_WIDTH, 1], which='both')
        ax2.set_position([.5, 0, DISC_WIDTH, 1], which='both')

        if i == len(g_vals) - 1:
            subfig.suptitle(fr'$\nu = {nu}$', y=0)

        if j == 0:
            ax1.set_ylabel(fr'$g = {g}$')

# plt.subplots_adjust(bottom=0, top=1, left=.05, right=.95, wspace=0)
# fig.subplots_adjust(bottom=0, top=1, left=0, right=1, wspace=.1, hspace=.1)
```
