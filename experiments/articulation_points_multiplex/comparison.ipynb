```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
save_results = True
```

```python
with open('results/lambda_4_no_cascade/results_lambda_4_no_cascade.json', 'r') as json_file:
    results = json.load(json_file)
    N = results['N']
    nruns = results['nruns']
    crange_4 = np.linspace(*results['crange'])
    results_4 = results['results']
```

```python
n_ap_4 = []
for result in results_4:
    n_ap_loc = sum(run['n_ap'] for run in result['runs'])
    n_ap_4.append(n_ap_loc / N / nruns)
```

```python
with open('results/lambda_3/results_lambda_3.json', 'r') as json_file:
    results = json.load(json_file)
    N = results['N']
    nruns = results['nruns']
    crange_3 = np.linspace(*results['crange'])
    results_3 = results['results']
```

```python
with open('results/ER/results_ER.json', 'r') as json_file:
    results = json.load(json_file)
    N = results['N']
    nruns = results['nruns']
    crange_ER = np.linspace(*results['crange'])
    results_ER = results['results']
```

```python
n_ap_ER = []
for result in results_ER:
    n_ap_loc = sum(run['cascade'][0]['n_ap'] for run in result['runs'])
    n_ap_ER.append(n_ap_loc / N / nruns)
```

```python
lw, ms = 1, 3
```

```python
plt.plot(crange_3, n_ap_3, 'o-', lw=lw, ms=ms, color='tab:blue', label='Scale-free $\lambda = 3$')
plt.plot(crange_4, n_ap_4, 'o-', lw=lw, ms=ms, color='tab:red', label='Scale-free $\lambda = 4$')
plt.plot(crange_6, n_ap_6, 'o-', lw=lw, ms=ms, color='tab:green', label='Scale-free $\lambda = 6$')
plt.plot(crange_ER, n_ap_ER, 'o-', lw=lw, ms=ms, color='tab:orange', label='ErdsRnyi')
plt.xlim(0, 8)
plt.ylim(0, .4)
plt.legend()
plt.xlabel('$c$', fontdict={'size': 18})
plt.ylabel('$n_{AP}$', fontdict={'size': 18})
plt.tick_params(axis='both', which='major', labelsize=12)
if save_results:
    plt.savefig(f'results/model_comparison.png')
    plt.savefig(f'results/model_comparison.pdf')
```
